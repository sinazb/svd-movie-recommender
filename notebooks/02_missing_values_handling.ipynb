```python
import pandas as pd
import numpy as np

ratings = pd.read_csv(
    "../data/ml-100k/u.data",
    sep="\t",
    names=["user_id", "movie_id", "rating", "timestamp"]
)

user_movie = ratings.pivot(
    index="user_id",
    columns="movie_id",
    values="rating"
)
```

```python
user_means = user_movie.mean(axis=1)
user_means.head()
```

```python
user_movie_centered = user_movie.sub(user_means, axis=0)
```

```python
user_movie_filled = user_movie_centered.fillna(0)
user_movie_filled.head()
```
